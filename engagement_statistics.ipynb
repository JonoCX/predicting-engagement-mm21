### Engagment Statistics


```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 

mpl.rcParams['figure.figsize'] = (15, 10)
sns.set(style = 'ticks', font_scale = 3.0)
sns.set_palette(sns.color_palette('colorblind'))
```

```python
ues = pd.read_csv(
    'data/ues_responses.csv',
    usecols = ['user_id', 'FA-Score', 'AE-Score', 'PU-Score', 'RW-Score', 'UES-Score']
)

print(f"Number of users: {len(ues.index)}")
```

```python
ues[['FA-Score', 'AE-Score', 'PU-Score', 'RW-Score', 'UES-Score']].describe()
```

```python
p = sns.distplot(ues['UES-Score'], kde = False)
p.set(
    xlabel = 'User Engagement Score\n(M = {0}, STD = {1}, MED = {2})'.format(
        round(ues['UES-Score'].mean(), 2),
        round(ues['UES-Score'].std(), 2),
        round(ues['UES-Score'].median(), 2)
    ),
    ylabel = 'Number of Users (N = {0})'.format(len(ues)),
    xticks = [1, 2, 3, 4, 5]
)

sns.despine()
plt.tight_layout()
plt.show()
```

```python
p = sns.boxplot(
    data = ues[['UES-Score', 'FA-Score', 'AE-Score', 'PU-Score', 'RW-Score']],
    orient = 'h',
    width=0.5,
    flierprops = {'alpha': 0.5, 'markersize': 3}
)
p.xaxis.grid(True, alpha = 0.5, linestyle = 'dashed')
p.set_yticklabels([
    f"Engagement Score\n(M = {round(ues['UES-Score'].mean(), 2)}, STD = {round(ues['UES-Score'].std(), 2)})", 
    f"Focused Attention\n(M = {round(ues['FA-Score'].mean(), 2)}, STD = {round(ues['FA-Score'].std(), 2)})", 
    f"Aesthetic Appeal\n(M = {round(ues['AE-Score'].mean(), 2)}, STD = {round(ues['AE-Score'].std(), 2)})", 
    f"Perceived Usability\n(M = {round(ues['PU-Score'].mean(), 2)}, STD = {round(ues['PU-Score'].std(), 2)})", 
    f"Reward\n(M = {round(ues['RW-Score'].mean(), 2)}, STD = {round(ues['RW-Score'].std(), 2)})"
])

sns.despine()
plt.tight_layout()
plt.show()
```

**Plotting for PhD Thesis**

```python
mpl.rc_file_defaults()
sns.reset_defaults()

sns.set_context('paper')
```

```python
p = sns.boxplot(
    data = ues[['UES-Score', 'FA-Score', 'AE-Score', 'PU-Score', 'RW-Score']],
    orient = 'h',
    width=0.3,
    linewidth = 1.1, 
    notch = False,
    showfliers = False,
    palette = sns.color_palette('colorblind'),
)

p.set(
    xlabel = 'Score', 
    yticklabels = [
        'Engagement', 'Focused Attention', 'Aesthetic Appeal', 'Perceived Usability', 'Reward'
    ],
    xticks = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
)
p.xaxis.grid(True, alpha = 0.5, linestyle = 'dashed')
p.set_axisbelow(True)
sns.despine(trim = True)
plt.tight_layout()
plt.savefig('plots/click-ues-factor-distribution.png')
# plt.show()
```

```python
p = sns.distplot(
    ues['UES-Score'], 
    kde = False, 
    color = sns.color_palette('colorblind')[0],
    hist_kws = {'edgecolor': sns.color_palette('colorblind')[0], 'linewidth': 0.2}
)
p.set(
    xlabel = 'User Engagement Score',
    ylabel = f"Number of Users (N = {len(ues)})",
    xticks = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
)
p.set_rasterized(True)
sns.despine()
plt.tight_layout()
# plt.show()
plt.savefig('plots/click-ues-distribution.png')
# plt.savefig('click-ues-distribution.svg', format='svg', dpi=1200)
```
